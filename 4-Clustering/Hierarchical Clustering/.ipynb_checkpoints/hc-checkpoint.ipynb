# Clustering Jerárquico

# Cómo importar las librerías


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Importar el data set


```python
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:, [3, 4]].values # .values para extraer los valores
# Seguimos con el problema de segmentar nuestros clientes que utilizan la tarjeta para compras en el shoping, utilizando el ingreso anual y su índice de gasto (mientras más chico menos gasta)
# Nuestra misión sigue siendo segmentar al conjunto de clientes según características que consideremos imprescindibles
```

# Utilizar el dendrograma para encontrar el número óptimo de clusters

```python
import scipy.cluster.hierarchy as sch # importamos para usar el dendrograma
dendrogram = sch.dendrogram(sch.linkage(X, method = "ward")) # construimos el dendrograma, linkage para indicarle que usamos CJ Aglomerativo, sobre la matriz de características X y el método Ward, que intenta minimizar la varianza entre los puntos de cada uno de los clusters (intracluster)
plt.title("Dendrograma")
plt.xlabel("Clientes") # puntos del dataset
plt.ylabel("Distancia Euclídea") 
plt.show()

# Debemos buscar el número óptimo de clusters identificando la distancia vertical más grande que no posea/cruce en su tramo ninguna otra recta horizontal, para cortar el dendrograma en cualquier parte de su recorrido, contando el número de lineas verticales que interceptamos para saber el número de clusters óptimo
# Recordar que podemos también tomar tramos de las lineas verticales, siempre que sean las más largas
# En CJ Aglomerativo, vemos que la linea más larga es la última azul de la derecha, pero a partir del tramo que supera a la 2da linea horizontal más alta amarilla hasta el techo rojo
# Si cortamos en el tramo reconocido, podemos ver que interceptaríamos 5 lineas verticales, por lo que ese sería nuestro número óptimo de clusters
# Recordemos que en K-Means también obtuvumos 5 clusters
```

# Ajustar el clustetring jerárquico a nuestro conjunto de datos

```python
from sklearn.cluster import AgglomerativeClustering # utilizamos la técnica de CJ aglomerativa, la más normal
hc = AgglomerativeClustering(n_clusters = 5, metric = "euclidean", linkage = "ward") # utilizamos el número óptimo de clusters, la distancia euclidea y el método de enlace para unir los clusters que se va a usar, respetando la manera en que creamos el dendrograma con ward (minimiza la varianza dentro del cluster) (podemos variar a los enfoques del punto más cercano, lejano, valor medio o punto geométrico medio del cluster, pero si lo cambiamos debemos aplicarlo también primero en el dendrograma)
y_hc = hc.fit_predict(X) # le decimos que ajuste y prediga los clusters de cada cliente
# Si visualizamos y_hc podemos ver en que cluster ubica a cada cliente concreto
```

# Visualización de los clusters

```python
# Si los datos son en 2D, lo más interesante es ver como se segmentan los datos en un gráfico (sino usar Técnicas de Reducción de Dimensionalidad)
plt.scatter(X[y_hc == 0, 0], X[y_hc == 0, 1], s = 100, c = "red", label = "Tacaños (Cluster 1)")
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1, 1], s = 100, c = "blue", label = "Estándard (Cluster 2)")
plt.scatter(X[y_hc == 2, 0], X[y_hc == 2, 1], s = 100, c = "green", label = "Millonarios (Cluster 3)") # campaña de márketing dirigida a estos, que son los que más gastan
plt.scatter(X[y_hc == 3, 0], X[y_hc == 3, 1], s = 100, c = "cyan", label = "Descuidados (Cluster 4)")
plt.scatter(X[y_hc == 4, 0], X[y_hc == 4, 1], s = 100, c = "magenta", label = "Cautos (Cluster 5)")
plt.title("Cluster de clientes")
plt.xlabel("Ingresos anuales (en miles de $)")
plt.ylabel("Puntuación de Gastos (1-100)")
plt.legend()
plt.show()

# Vemos que cambia la segmentación de algunos datos, particularmente concibe un par de puntos más como clientes Estándard con respecto a K-Means
# En CJ no nos sirve visulizar los baricentros, ya que no se calculan en ningún momento ni se utilizan para segmentar los datos, como si pasa en K-Means
# El análisis de los clusters es el mismo que en K-Means
```
