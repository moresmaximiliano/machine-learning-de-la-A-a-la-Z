# K-Means

# Cómo importar las librerías


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import os
os.environ['OMP_NUM_THREADS'] = '1' # para eliminar alertas por fuga de memoria
```

# Importar el data set


```python
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:, [3,4]].values # Incluimos solo los ingresos anuales y coeficiente de su gasto en centro comercial
# El dataset nos muestra información de clientes de un centro comercial que se suscribe a una tarjeta de crédito, con una puntuación de cuanto gasta cada usuario en el centro comercial (números bajos gastan poco con tarjeta, números altos mucho)
# Queremos segmentar los clientes en base a sus ingresos anuales y su proporción de gasto, para caracterizarlos
```

# Método del codo para averiguar el número óptimo de clusters

```python
# Visualizamos el valor de WCSS para cada número de cluster (del 1 al 10) para encontrar el cambio de tendencia que produce el número de cluster óptimo (la última disminución brusca del WCSS) 
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11): # del 1 al 10
    kmeans = KMeans(n_clusters = i, init = "k-means++", max_iter = 300, n_init = 10, random_state = 0) # n_clusters: cuántos clusters queremos, init: inicialización de los varicentros, como no queremos caer en la trampa de la iniciación aleatoria usamos k-means ++, max_iter: como podría ser que el algoritmo nunca acabara (redefiniendo eternamente los baricentros), le asignamos un número máximo de iteraciones para redefinirlo (osea de ajustes del baricentro, por defecto 300), n_init: inicialización aleatoria, por defecto 10
    kmeans.fit(X)
    wcss.append(kmeans.inertia_) # agregamos a la lista el valor de la suma de los cuadrados de la distancia de cada punto a su baricentro para los distintos números de clusters

plt.plot(range(1,11), wcss)
plt.title("Método del codo") 
plt.xlabel("Número de Clusters")
plt.ylabel("WCSS(k)")
plt.show()

# Identificar la última disminución brusca del WCSS para establecer el número óptimo de clusters. Vemos que el número en donde se produce el cambio de tendencia es en el número de cluster igual a 5 (última disminución brusca)
```

# Aplicar el método de k-means para segmentar el data set

```python
# Creamos el objeto de kmeans pero ya conociendo el número óptimo de clusters a identificar
kmeans = KMeans(n_clusters = 5, init="k-means++", max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(X) # aplicamos entrenamiento y predicción, para que me segmente directamente a las observaciones

# Recordar que el Clustering se aplica sobre todos los datos sobre los cuales queremos segmentar, reconocidos como X y sin aplicar previamente una división en train/test
```

# Visualización de los clusters

```python
# Si los datos son en 2D, lo más interesante es ver como se segmentan los datos en un gráfico (sino usar Técnicas de Reducción de Dimensionalidad)
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = "red", label = "Estándard (Cluster 1)") # pintar los valores según pertenezcan al cluster 0
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = "blue", label = "Millonarios (Cluster 2)")
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = "green", label = "Tacaños (Cluster 3)")
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s = 100, c = "cyan", label = "Cautos (Cluster 4)")
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], s = 100, c = "magenta", label = "Descuidados (Cluster 5)")
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s = 300, c = "yellow", label = "Baricentros") # pintamos el centro geométrico de cada cluster
plt.title("Cluster de clientes")
plt.xlabel("Ingresos anuales (en miles de $)")
plt.ylabel("Puntuación de Gastos (1-100)") # los colocamos así porque suponemos que el gasto(y) se produce en función del ingreso(x), y no viceversa (también ese es el orden establecido por el set)
plt.legend()
plt.show()

# Luego toca interpretar que tipo de clientes puede pertenecer a cada cluster, concluyendo que:
# - Los de ingreso medio y gasto medio son Estándard
# - Los de ingreso alto y gasto alto son Millonarios (objetivo probable de una campaña de Marketing, suponiendo que no alteran su comportamiento)
# - Los de ingreso alto y gasto bajo son Tacaños 
# - Los de ingreso bajo y gasto alto son Descuidados
# - Los de ingreso bajo y gasto bajo son Cautos

# Podemos ver en y_kmeans la pertenencia al cluster de cada dato
# Si tomamos solo 2 características según las cuales clusterizar, podemos visualizar el gráfico
```
