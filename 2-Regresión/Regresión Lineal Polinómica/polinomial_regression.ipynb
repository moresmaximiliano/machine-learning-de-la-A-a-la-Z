# Regresión polinómica


# Cómo importar las librerías


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Importar el data set


```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:2].values # solo tomamos las categorizaciones numéricas de los puestos, lo hacemos de este modo para que lo tome como matriz (y no vector) de características correctamente (2da columna) (pasa de ser de una dimensión de (10,) con iloc[:,1] a una de (10,1) con este iloc)
y = dataset.iloc[:, 2].values # tomamos los sueldos 
# Como la variable de los puestos si es ordinal, no debemos aplicar One-Hot Encoder
```

```python
# Si vemos como se distribuyen los datos con un scatter, observamos que le distribución de los datos tiende a ser exponencial, no lineal
# Creamos el scatter plot
plt.scatter(X, y, color='red') # si tenemos varias variables lo podemos hacercon cada una en particular con respecto a y
plt.title('Salarios en función de la Posición')
plt.xlabel('Posición Nivel')
plt.ylabel('Salario')
plt.show()
```

# Ajustar la regresión lineal con el dataset

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,y) # como no tenemos x_train ni y_train seleccionamos todo para entrenar
```

# Ajustar la regresión lineal polinómica con el dataset

```python
from sklearn.preprocessing import PolynomialFeatures
# Preparamos la transformación de nuestra matriz de características especificando hasta que grado queremos la transformación polinómica (potencias grado 4 de X)
poly_reg = PolynomialFeatures(degree = 4) # vemos que mientras más grados le agregamos a la matriz, más mejora el entrenamiento de la variable y el grado de ajuste a los datos de entrenamiento
# Transformamos los valores de las variables con valores polinómicos para mantener la linealidad de los coeficientes que luego serán calculados con la regresión lineal
X_poly = poly_reg.fit_transform(X) # me agrega la columna de la v.indep, la X base, X^2, X^3 y X^4 (si ponemos degree=4, si ponemos =2 solo hasta la X^2)
# Hacemos la regresión lineal con esa matriz de características polinómica (X_poly). Tendremos 4 coeficientes 
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)

for i in range (len(X_poly)):
    print(X_poly[i].tolist())

lin_reg_2.get_params() # para ver los parámetros de la regresión hecha
lin_reg_2.score(X_poly,y) # vemos el R2 de la predicción de la regresión (con datos de entrenamiento)
```

# Visualización de los resultados del Modelo Lineal

```python
# Obtener los coeficientes de la ecuación de regresión
intercept = lin_reg.intercept_
coef = lin_reg.coef_

print(f"Ecuación de regresión: y = {intercept} + {coef[0]} * X")
```

```python
plt.scatter(X, y, color = "red")
plt.plot(X, lin_reg.predict(X), color = "blue")
plt.title("Modelo de Regresión Lineal")
plt.xlabel("Posición del empleado")
plt.ylabel("Sueldo (en $)")
plt.show()

# Empleado en posición 1 tendría sueldo negativo
# Vemos que el modelo lineal infravalora las primeras 2 posiciones, explica bien el punto de la 3ra, luego sobrevalora hasta la 9na posición e infravalora en una gran cuantía la 10ma, por lo que no se adecua mucho a los datos (salvo para la posición 2 y 9), siendo conveniente rechazarlo
```

# Visualización de los resultados del Modelo Polinómico

```python
# Obtener los coeficientes de la ecuación de regresión
intercept = lin_reg_2.intercept_
coef = lin_reg_2.coef_
print(f"Ecuación de regresión: y = {intercept} + {coef[1]} * X + {coef[2]} * X^2 + {coef[3]} * X^3 + {coef[4]} * X^4") # empezamos de [1] xq está la columna con 1
```

```python
# Crear X_grid para una visualización más suave
X_grid = np.arange(min(X.flatten()), max(X.flatten()), 0.1) # creamos un vector de datos intermedios de X (entre cada valor de X por 0.1, mejorando la visualización en el gráfico) y también pido que me aplane el array 
X_grid = X_grid.reshape(len(X_grid), 1) # ponemos para cambiar la dimensión de los datos

# Crear el scatter plot y la línea de regresión polinómica
plt.scatter(X, y, color='red')
plt.plot(X_grid, lin_reg_2.predict(poly_reg.fit_transform(X_grid)), color='blue') # le pasamos los datos con las nuevas columnas requeridas (elevadas)
plt.title('Modelo de Regresión Polinómica')
plt.xlabel('Posición del empleado')
plt.ylabel('Sueldo (en $)')
plt.show()

# Si hacemos la regresión polinómica de grado 2, nos aparece una curva que decrece en un punto, pero a medida que vamos aumentando los grados mejora el ajuste a los datos (llega a un punto máximo)
# En grado 2 en convexo, en grado 3 aparece un punto de inflexión y en grado 4 ya desaparece una disminución a medida que aumentamos de puesto
```

# Predicción de nuestros modelos

```python
# Nivel operativo del futuro empleado: 6,5 (entre ambas categorias)
# Elaboramos la predicción con el modelo lineal de un sueldo para un nivel operativo de 6.5
print(lin_reg.predict([[6.5]])) 
# Elaboramos la predicción con el modelo polinómico de un sueldo para un nivel operativo de 6.5
print(lin_reg_2.predict(poly_reg.fit_transform([[6.5]])))

# Con la predicción de la regresión lineal perderíamos mucha plata como empresa, a comparación del polinomial, que nos ayudará a contratar a buenos sueldos
```
