```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Importar el data set


```python
dataset = pd.read_csv('50_Startups.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 4].values # 5ta columna
# Son 50 empresas con datos financieros (gasto en I+D, en Administración y en Marketing) y geográficos (donde está ubicada la empresa) de las mismas, en conjunto con su correspondiente beneficio generado anual(y)
# 4 variables indep (3 numéricas y 1 categórica) y 1 dependiente (beneficios)
# Como no poseemos información acerca de qué empresas se trata, buscamos hacer una regresión para predecir con esos datos sobre los beneficios de nuevas empresas, en vez de realizar una regresión múltiple y ver que empresas estan subvaluadas y sobrevaloradas (comparar cada y_pred del modelo fiable con su y_test)
# Buscamos ver con los coeficientes de regresión que gastos son los que me indican en mayor medida sobre el futuro beneficio de la empresa
```

```python
X # vemos que ahora tenemos la columna de 1 al principio. El coeficiente que se lleve la primera columna será el termino independiente, pudiendo medir su p-valor al igual que las variables numéricas (sin que importe si es mayor a 0.05)
```

# Automatizamos el proceso de Eliminación Hacia Atras

## Eliminación hacia atrás utilizando solamente p-valores

```python
import numpy as np
import statsmodels.api as sm

def eliminacionHaciaAtras(x, y, sl):
    x = np.asarray(x, dtype=np.float64)  # Asegurar que x sea un array numpy de tipo float
    y = np.asarray(y, dtype=np.float64)  # Asegurar que y sea un array numpy de tipo float
    numVars = x.shape[1]
    selected_vars = list(range(numVars))  # Lista de índices de variables seleccionadas
    
    for i in range(numVars):
        regressor_OLS = sm.OLS(y, x).fit()
        maxVar = max(regressor_OLS.pvalues)
        if maxVar > sl:
            for j in range(x.shape[1]):
                if regressor_OLS.pvalues[j] == maxVar:
                    x = np.delete(x, j, 1)
                    del selected_vars[j]  # Eliminar el índice de la variable
                    break
    print(regressor_OLS.summary()) # imprimo salida de la regresión final
    return x, selected_vars

# Usamos la función con nuestros datos
SL = 0.05
X_opt = X[:, [0, 1, 2, 3, 4, 5]]
X_Modeled, selected_vars = eliminacionHaciaAtras(X_opt, y, SL)

print('--------------------------------------------------------------------------------')
print('--------------------------------------------------------------------------------')
print("Variables seleccionadas:", selected_vars) # me muestra que aplicando el método rigurosamente, me tengo que quedar con X3 solo
# Este método según el p-valor puede ser un poco arbitrario al momento de descartar variables
```

## Eliminación hacia atrás utilizando  p-valores y el valor de  R Cuadrado Ajustado

```python
import numpy as np
import statsmodels.api as sm

def eliminacionHaciaAtras(x, y, SL):
    # Asegurar que x e y sean arrays numpy de tipo float64
    x = np.asarray(x, dtype=np.float64)
    y = np.asarray(y, dtype=np.float64)

    numVars = x.shape[1]
    temp = np.zeros((x.shape[0], numVars))
    selected_vars = list(range(numVars))  # Lista de índices de variables seleccionadas
    for i in range(numVars):
        regressor_OLS = sm.OLS(y, x).fit()
        maxVar = max(regressor_OLS.pvalues)
        adjR_before = regressor_OLS.rsquared_adj
        if maxVar > SL:
            for j in range(x.shape[1]):
                if regressor_OLS.pvalues[j] == maxVar:
                    temp[:, j] = x[:, j]
                    x = np.delete(x, j, 1)
                    tmp_regressor = sm.OLS(y, x).fit()
                    adjR_after = tmp_regressor.rsquared_adj
                    if adjR_before >= adjR_after:
                        x = np.hstack((x, temp[:, [j]].reshape(-1, 1)))
                        x = np.delete(x, j, 1)
                        print(regressor_OLS.summary())
                        return x, selected_vars
                    else:
                        del selected_vars[j]  # Eliminar el índice de la variable
                        continue
    print(regressor_OLS.summary()) # imprimo salida de la regresión final
    return x, selected_vars

# Usamos la función con nuestros datos
SL = 0.05
X_opt = X[:, [0, 1, 2, 3, 4, 5]]
X_Modeled, selected_vars = eliminacionHaciaAtras(X_opt, y, SL)
print('--------------------------------------------------------------------------------')
print('--------------------------------------------------------------------------------')
print("Variables seleccionadas:", selected_vars) # me muestra que priorizando el R2Adj me tengo que quedar con X3 y X5 (la última eliminación de p-valor me quita R2 Adj)
# Este criterio sirve para comparar modelos con diferentes cantidades de VI (si estamos comparando con un modelo simple tomar para comparar su R2)
# Vemos que el modelo final me arroja los siguientes resultados: coeficientes positivos(+) en Gastos de I+D(0,79) y Marketing(0,03) para explicar las ganancias, y la empresa generaría 46980 sin gastos (ya que es la o.o y su p-valor<0.05)
# La magnitud de la inlfuencia en I+D es muy superior al de Marketing (pero prestar atención a las unidades en que está medida cada una, si ambas estan en dolares podemos afirmar lo anterior, pero si Marketing esta medida en centavos no)
# Como estan las 2 en dolares, por cada dolar gastado en I+D, obtengo un 0,79 de ganancia y por cada dolar gastado en Marketing, obtengo 0,03 de ganancia
```
